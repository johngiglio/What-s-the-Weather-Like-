```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import datetime
# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
np.random.seed(123)
lat_lngs = []
cities = []
geo_data = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1])
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        geo_data.append({'City':city.city_name.title(),'Country':city.country_code.upper(), 'Lat':lat_lng[0], 'Lon':lat_lng[1]})
geo_df = pd.DataFrame(geo_data)


# Print the city count to confirm sufficient count
len(geo_df)
```

```python
lat_bins = list(range(-90,91,10))
lon_bins = list(range(-180,181,20))
geo_df['Lat Bin'] = pd.cut(geo_df['Lat'],lat_bins)
geo_df['Lon Bin'] = pd.cut(geo_df['Lon'], lon_bins)
```

```python
lat_group = pd.DataFrame(geo_df.groupby('Lat Bin')['City'].count())
lat_group = lat_group.rename(columns = {'City':'# of Cities'})
```

## Perform API Calls

Now that the city list has been built, the OpenWeatherMap API will be used to retrieve weather data for each city. Store temp, humidity, and other key metrics for each city. 

```python
#Initialize lists to store weather results as dictionaries, and info about cities not found
weather_results = []
cities_not_found = []
counter = 1

# Starting URL for Weather Map API Call

for i in range(len(geo_df)):
    url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={api_key}&q={geo_df['City'][i]},{geo_df['Country'][i]}"
    #OpenWeatherAPI does not have all cities
    try:
        resp = requests.get(url).json()
        weather_results.append({'City':geo_df['City'][i],
                                'Country':geo_df['Country'][i],
                                'Latitude':resp['coord']['lat'],
                                'Longitude':resp['coord']['lon'],
                                'Temperature':resp['main']['temp'],
                                'Minimum Temperature (F)':resp['main']['temp_min'],
                                'Maximum Temperature (F)':resp['main']['temp_max'],
                                'Humidity (%)':resp['main']['humidity'],
                                'Cloudiness (%)': resp['clouds']['all'],
                                'Wind Speed (mph)':resp['wind']['speed'],
                                'Sunrise':resp['sys']['sunrise'],
                                'Sunset':resp['sys']['sunset']})
        
        print(f"City #{counter}: {geo_df['City'][i]},{geo_df['Country'][i]}")
        counter += 1
    except:
        cities_not_found.append([geo_df['City'][i],geo_df['Country'][i]])
        print(f"{geo_df['City'][i]},{geo_df['Country'][i]} Not Found!")
        
    time.sleep(0.05)
    # json.dumps(resp)
```

The results below were obtained on July 21, 2018 at approximately 8:00 PM EDT/12:00 AM UTC. Because the API returns only the current temperature, and not a max/min/average temperature for the day, it may be important to consider the time of day for each city to observe the effect of sunlight at the time of the measurement. The time of day for each location will be calculated by approximating that each 15 degree change in longitude from the prime meridian corresponds to a 1 hour difference in time from the UTC time zone. 

```python
current_weather = pd.DataFrame(weather_results)
current_weather.head()
```

```python
current_weather['Hour'] = np.where(round(current_weather['Longitude']/15,0) < 0,
                                   (round(24 + current_weather['Longitude']/15,0).astype(int)), 
                                   (round(current_weather['Longitude']/15,0).astype(int)))
current_weather.head()
```

## Temperature vs. Latitude

The figure below shows the temperature vs. latitude for each city. Unsurprisingly, we see a roughly parabolic shape in this plot, where the temepratures are higher near 0 latitude and lower near the edges of the figure. The unexpected result of this plot is the lack of relationship that time of day (day/night) has on the temperature. For the most part Day and Night data points have similar distributions.

```python
fig, ax = plt.subplots(figsize = (20,10))

colors = {'Day':'yellow','Night':'black'}
for i in colors:
    ax.scatter(current_weather[current_weather['Day/Night'] == i]['Latitude'],current_weather[current_weather['Day/Night']== i]['Temperature'],alpha = 0.5, s = 100,c =colors[i])



#plt.scatter(current_weather['Latitude'],current_weather['Temperature'],alpha = 0.5, s = 100,c =[colors[i] for i in current_weather['Day/Night']])
ax.set_title('Temperature vs. Latitiude on 7/21/18', fontsize = 25)
ax.set_xlabel('Latitude',fontsize=15)
ax.set_ylabel('Temperature (F)',fontsize=15)
ax.legend(['Day','Night'],loc = 'best')
fig.tight_layout()
fig.savefig('tempvslat.png')
plt.show()
```

## Cloudiness vs. Latitude

The plot below shows cloudiness vs. latitude for each city. Like the previous two plots, this plot does not show an obvious relationship between cloudiness and latitude. Because cloudiness % is stored as an integer and is bounded by the range 0-100, many of the cities in the dataframe have exactly the same value for cloudiness. The clustering of points with the same cloudiness appear to form horizontal lines across the figure, clearly demonstrating that cities at very different latitudes experienced the same amount of cloudiness at the time that measurements were recorded.

```python
fig, ax = plt.subplots(figsize = (20,10))
plt.scatter(current_weather['Latitude'],current_weather['Cloudiness (%)'])
ax.set_title('Cloudiness vs. Latitiude', fontsize = 25)
ax.set_xlabel('Latitude',fontsize=15)
ax.set_ylabel('Cloudiness (%)',fontsize=15)
fig.savefig('cloudvslat.png')
plt.show()
```

## Conclusion

Latitude appears to have the greatest effect on temperature. A subset of the data in this analysis appeared to show that humidity will generally be higher near the equator, but otherwise is not affected by latitude. Latitude alone does not appear to have much of an effect on cloudiness or wind speed. Some limitations of this data set are that the exact local time for each city could not be determined. This anaylsis would benefit from a mroe granular assessment of the effects of time of day/sunlight on temperature.
